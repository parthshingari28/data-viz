```python
from matplotlib import pyplot as plt
```

```python
plt.style.use("fivethirtyeight")
```

```python
slices = [60, 40]

plt.pie(slices)
```

```python
slices = [60, 40]
# values need not to be complete 100%
labels = ['Sixty', 'Forty']

plt.pie(slices, labels = labels, wedgeprops = {'edgecolor': 'black'})

plt.title("My awesome Pie Chart")

plt.savefig("Awesome-Pie-Chart.png")
```

```python
slices = [60, 40, 10, 35, 7]

labels = ['CSK', 'MI', 'RR', 'RCB', 'KKR']

colors = ['yellow', 'blue', 'pink', 'red', 'purple']

plt.pie(slices, labels = labels, colors = colors, wedgeprops = {'edgecolor':'black'})

plt.title("IPL - 2021")

plt.savefig("IPL-2021.png")
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels = labels, colors = colors, wedgeprops = {'edgecolor':'black'})

plt.title("Programming Language Popularity")

plt.savefig("Language-Popularity-Chart.png")
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

plt.pie(slices, labels = labels, wedgeprops = {'edgecolor':'black'})

plt.title("Top 5 - Programming Language Popularity")

plt.savefig("Top-5 Langauge Popularity.png")
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
# suppose, here we wanted to explode python and js
explode = [0.1, 0, 0, 0.1, 0]

plt.pie(slices, labels = labels,explode = explode, wedgeprops = {'edgecolor':'black'})

plt.title("Top 5 - Programming Language Popularity")

plt.savefig("Langauge-Popularity-PieChart.png")
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
# suppose, here we wanted to explode python and js
explode = [0.1, 0, 0, 0.1, 0]

# add shadow = true -> makes chart look bit of 3D type
plt.pie(slices, labels = labels,explode = explode, shadow = True, wedgeprops = {'edgecolor':'black'})

plt.title("Top 5 - Programming Language Popularity")

plt.savefig("Langauge-Popularity-PieChart-3D.png")
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
# suppose, here we wanted to explode python and js
explode = [0.1, 0, 0, 0.1, 0]

# display percentage -> autopct method
# add shadow = true -> makes chart look bit of 3D type
plt.pie(slices, labels = labels,explode = explode, shadow = True, 
        autopct = '%1.1f%%',
        wedgeprops = {'edgecolor':'black'})

plt.title("Top 5 - Programming Language Popularity")

plt.savefig("Langauge-Popularity-PieChart-v2.png")
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
# suppose, here we wanted to explode python and js
explode = [0.1, 0, 0, 0.1, 0]

# startangle method
# add shadow = true -> makes chart look bit of 3D type
plt.pie(slices, labels = labels,explode = explode, shadow = True,
        startangle = 90,
        wedgeprops = {'edgecolor':'black'})

plt.title("Top 5 - Programming Language Popularity")

plt.savefig("Langauge-Popularity-PieChart-v3.png")
```
